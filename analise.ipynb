Taxas de Rendimento Escolar (Aprovação, Reprovação e Abandono), segundo a Localização e a Dependência Administrativa, nos Níveis de Ensino Fundamental e Médio - Brasil, Regiões Geográficas e Unidades da Federação - 2020

## Carregando o conjunto de dados:

```python
import pandas as pd

tx = pd.read_excel('datasets/tx_rend_brasil_regioes_ufs_2020.xlsx', skiprows=8)
tx
```

```python
tx.info()
```

## Preparando os dados para Análise

**Obs.:** Todas as colunas estão sendo identificadas como Object

Antes de resolver esses problemas vamos solucionar as colunas Unidade Geográfica, Localização, Dependência Administrativa, Total de taxa de Aprovação: Ensino fundamental de 8 a 9 anos e Total de Ensino Médio, Total de Taxa de Reprovação: Ensino fundamental de 8 a 9 anos e Total de Ensino Médio e Total de taxa de Abandono: Ensino fundamental de 8 a 9 anos e Total de Ensino Médio para efetuar a análise.

```python
tx = tx.iloc[:,[1, 2, 3, 4, 16,  22, 34, 40, 52]]
tx
```

Dicionário:
- UG = Unidade Geográfica
- LOC = Localização
- DA = Dependência Administrativa
- TTAEF = Taxa Total de Aprovacão Ensino Fundamental
- TTAEM = Taxa Total de Aprovacão Ensino Médio
- TTREF = Taxa Total de Reprovação Ensino Fundamental
- TTREM = Taxa Total de Reprovação Ensino Médio
- TTABEF = Taxa Total de Abandono Ensino Fundamental
- TTABEM = Taxa Total de Abandono Ensino Médio

```python
colunas = ['UG', 'LOC', 'DA', 'TTAEF', 'TTAEM', 'TTREF', 'TTREM', 'TTABEF', 'TTABEM']
tx.columns = colunas
tx
```

```python
tx.info()
```

Algumas colunas estão com valores não definidos ou cujo valor não é conhecido corretamente, mas isso não significa que não existam valores. Irei substituir o '--' por NaN (Not a Number), pois o pandas lidará com os dados ausentes corretamente, sem distorcer os cálculos.

```python
import numpy as np

# Substitui os valores '--' por NaN, onde NaN vai prencher os valores ausentes com 0, sem alterar os dados das colunas.
df = tx.replace('--', np.nan)
```

```python

# Aqui percebemos que as colunas que estavam como 'object', agora estão como 'float64'.
df.info()
```

## Visualização dos dados
